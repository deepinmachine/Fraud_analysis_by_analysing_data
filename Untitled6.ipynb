```python
sku_master=pd.read_excel('/content/Company X - SKU Master.xlsx')
sku_master.info()
```

```python
merged_data = company_order_x.merge(sku_master, on='SKU', how='left')
```

```python
merged_data
```

```python
#we will calculate the total waeight and price of a order by multiplying Qty with item price and weight
# then we will group by by each order id to extract the price ans weight of each product
merged_data['total weight']=merged_data['Order Qty']*merged_data['Weight (g)']
merged_data['total_price_x']=merged_data['Order Qty']*merged_data['Item Price(Per Qty.)']
```

```python
#grouping data based on order id
grouped_data=merged_data.groupby('ExternOrderNo').agg({'total weight':'sum','total_price_x':'sum'}).reset_index()
grouped_data['total_weight(kg)_x']=grouped_data['total weight']/1000
grouped_data.info()
```

```python
grouped_data.drop('total weight', axis=1,inplace=True)
grouped_data.rename(columns={'ExternOrderNo':'Order ID'},inplace=True)
```

```python
grouped_data.describe()
```

```python
#now we will go for to differentiatie between zones courier and company
# 1. we will join pinocde zones data to coutier invoice this will create a new column zones_x
pincode=pd.read_excel('/content/Company X - Pincode Zones.xlsx')
courier_invoice=pd.read_excel('/content/Courier Company - Invoice.xlsx')
pincode.rename(columns={'Zone':'zone_x'},inplace=True)
```

```python
pincode.info()
```

```python
courier_invoice.info()
```

it is necessary to delete all the duplicates from the pincode table else it will create the large number of rowa

```python
pincode=pincode.drop_duplicates()
```

```python
#joining two tables
new_courier_invoice=courier_invoice.merge(pincode,on=['Customer Pincode'],how='left')
new_courier_invoice.info()
```

```python
#now we have price of item and weight of item and zone of item as per company x
#now we will have to find price of item as per the company x
# now we will have to find price according to weight slabs
#we will joins weight slabs to the new_courier_invoice on zone_x column so that we will have rates as per zone
```

```python
courier_rates=pd.read_excel('/content/Courier Company - Rates.xlsx')
courier_rates['Zone']=courier_rates['Zone'].str.lower()
courier_rates
```

```python
new_data=new_courier_invoice.merge(courier_rates,left_on='zone_x',right_on='Zone',how='left')
```

```python
new_data.head()
```

```python
#now we have to get weight that we calculate earlier to be joined on this new_data table
new_data.info()
```

```python
new_data1=new_data.merge(grouped_data,on='Order ID',how='inner')
```

```python
new_data1.info()
```

```python
new_data1.head()
```

```python
import math
```

```python
def charges(x):
  return len(x)
new_data1['Type of Shipment']=new_data1['Type of Shipment'].apply(charges)
```

```python
new_data1['Type of Shipment']=new_data1['Type of Shipment'].astype(int)
```

```python
# so now forward charge will be length 15 and forward and rto charges will be length 23
def calculation_price(new_data1):
  x=0
  price=0
  if new_data1['total_weight(kg)_x']>new_data1['Weight Slabs']:
    x=math.ceil(new_data1['total_weight(kg)_x']/new_data1['Weight Slabs'])
  else:
    x=1
  if new_data1['Type of Shipment']==15:
    price=new_data1['Forward Fixed Charge']+(x-1)*new_data1['Forward Additional Weight Slab Charge']
  else:
    price=new_data1['Forward Fixed Charge']+(x-1)*new_data1['Forward Additional Weight Slab Charge']+new_data1['RTO Fixed Charge']+(x-1)*new_data1['RTO Additional Weight Slab Charge']
  return price
```

```python
new_data1['price_x']=new_data1.apply(calculation_price,axis=1)
```

```python
#now renaming columns to as per given assignment
new_data1.rename(columns={'total_weight(kg)_x':'Total weight as per X (KG)','Weight Slabs':'Weight slab as per X (KG)','Charged Weight':'Total weight as per Courier Company (KG)','zone_x':'Delivery Zone as per X','Zone_x':'Delivery Zone charged by Courier Company','price_x':'Expected Charge as per X (Rs.)','Billing Amount (Rs.)':'Charges Billed by Courier Company (Rs.)'},inplace=True)
```

```python
new_data1['Difference Between Expected Charges and Billed Charges (Rs.)']=new_data1['Expected Charge as per X (Rs.)']-new_data1['Charges Billed by Courier Company (Rs.)']
```
